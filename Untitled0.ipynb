```python
def missing_percentage(df):
  # Calculate the percentage of missing values
  percent_missing = df.isnull().sum() * 100 / len(df)

  # Create a Series with column names and percentages
  missing_value_df = pd.DataFrame({'percent_missing': percent_missing})
  # Sort the results
  missing_value_df.sort_values('percent_missing', inplace=True, ascending = False)

  pd.set_option("display.max_rows", None)  # Set to None to display all rows


  return missing_value_df
```

```python
import pandas as pd

def find_binary_categorical_vars(df):

    binary_categorical_vars = []
    Multicategorical_vars = []
    for col in df.columns:
        if df[col].dtype == 'object' or pd.api.types.is_categorical_dtype(df[col]):  # Check if categorical
            unique_values = df[col].nunique()
            if unique_values == 2:  # Check if binary
                binary_categorical_vars.append(col)
            elif unique_values > 2:
              Multicategorical_vars.append(col)

    return binary_categorical_vars, Multicategorical_vars


binary_vars, multinomial_vars = find_binary_categorical_vars(df_train)
print("Binary categorical variables:", binary_vars)
print("Multinomial categorical variables:", multinomial_vars)
```

```python
def plotData(data, variable):

    plt.figure(figsize=(10, 6))  # Adjust figure size if needed

    if pd.api.types.is_numeric_dtype(data[variable]):
        # Numerical variable: Use histplot with kde
        sns.histplot(data[variable], kde=True)
        plt.title(f'Distribution of {variable}')
        plt.xlabel(variable)
        plt.ylabel('Frequency')
    else:
        # Categorical variable: Use countplot
        sns.countplot(x=variable, data=data, palette='hls')  # Use palette if desired
        plt.title(f'Distribution of {variable}')
        plt.xlabel(variable)
        plt.ylabel('Count')
        plt.xticks(rotation=45)  # Rotate x-axis labels if needed

    plt.show()
```

```python
def PlotStats(X, Y):
  plt.figure(figsize=(8, 6))
  sns.boxplot(x=X, y=Y, data=Model_df, palette='hls')

  # Calculate outliers
  Q1 = Model_df[Y].quantile(0.25)
  Q3 = Model_df[Y].quantile(0.75)
  IQR = Q3 - Q1
  outliers = Model_df[(Model_df[Y] < Q1 - 1.5 * IQR) | (Model_df[Y] > Q3 + 1.5 * IQR)]
  num_outliers = len(outliers)
  percent_outliers = (num_outliers / len(Model_df)) * 100

  # Update title to include outlier information
  plt.title(f'{X} vs {Y} Box Plot - Outliers: {num_outliers} ({percent_outliers:.2f}%)')

  if X == 'TARGET':
    plt.xlabel('Target (0: No Default, 1: Default)')
  else:
    plt.xlabel(f'{X}')
  plt.ylabel(f'{Y}')
  plt.xticks(rotation=45)

  plt.show()
```

```python
PlotStats(X = 'TARGET', Y = 'CNT_CHILDREN')
```

```python
def plot_countplots_by_target(data, variable, target_variable='TARGET'):
    # Figure 1: Defaults
    plt.figure(figsize=(10, 6))
    order = data[data[target_variable] == 1][variable].value_counts().index
    sns.countplot(
        x=variable,
        data=data[data[target_variable] == 1],
        color='red',
        order=order  # Add this line to sort bars
    )
    plt.title(f'Distribution of {variable} for Defaults')
    plt.xlabel(variable)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()

    # Figure 2: Non-Defaults
    plt.figure(figsize=(10, 6))
    order = data[data[target_variable] == 0][variable].value_counts().index
    sns.countplot(
        x=variable,
        data=data[data[target_variable] == 0],
        color='blue',
        order=order  # Add this line to sort bars
    )
    plt.title(f'Distribution of {variable} for Non-Defaults')
    plt.xlabel(variable)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
```

```python
plot_countplots_by_target(Model_df, 'NAME_EDUCATION_TYPE')
```

```python
plot_countplots_by_target(Model_df, 'OCCUPATION_TYPE')
```
